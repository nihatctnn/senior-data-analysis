```python
import pandas as pd;
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading the CSV file and assigning it to a DataFrame named 'df'

df = pd.read_csv(r"D:\windows11\desktop\senior-project\data-analysis\data-csv.csv")
df
```

```python
# Calculating the sum of missing values (null values) in each column of the DataFrame 'df'
df.isnull().sum()
```

```python
# Dropping rows with missing values from the DataFrame and checking for missing values again to confirm that they have been dropped.
df.dropna(axis=0, inplace=True)
df.isnull().sum()
```

```python
# Convert the 'Yıl' column to integer
df['Yıl'] = df['Yıl'].astype(int)

# general information about the dataset.
print(df.info())
```

```python
# Convert DataFrame to Excel file
# df.to_excel("data-xlsx.xlsx", index=False)
```

```python
most_recurring = df['Model'].mode()[0]
repetition_count = df['Model'].value_counts()[most_recurring]
print(f"The most recurring value in the 'model' column: {most_recurring}, Repetition count: {repetition_count}")
```

```python
filter = df['Model'] == most_recurring
most_recurring_data = df[filter]
# most_recurring_data.to_excel("most_recurring_data.xlsx", index=False)
print(f"Data with the most recurring value in the 'model' column has been saved to 'most_recurring_data.xlsx' file.")
```

```python
df_most_recurring = pd.read_excel('most_recurring_data.xlsx')
print(df_most_recurring)
```

```python
# Get year and price data
year = df_most_recurring['Yıl']
price = df_most_recurring['Fiyat']
# Plotting the graph
plt.figure(figsize=(10, 6))
plt.scatter(year, price, color='blue', alpha=0.7)
plt.title('Yıl-Fiyat Grafiği')
plt.xlabel('Yıl')
plt.ylabel('Fiyat')
plt.grid(True)
plt.show()
```

```python
# Get kilometer, price, and year data
km = df_most_recurring['Kilometre']
price = df_most_recurring['Fiyat']
year = df_most_recurring['Yıl']

# Get unique values in the Model column
unique_models  = df_most_recurring['Yıl'].unique()

# Assign different colors for each model
colors = plt.cm.rainbow(np.linspace(0, 1, len(unique_models)))

# Plotting the graph
plt.figure(figsize=(10, 6))
for i, model in enumerate(unique_models):
    filtre = df_most_recurring['Yıl'] == model
    plt.scatter(km[filtre], price[filtre], color=colors[i], alpha=0.7, label=model)

plt.title('Kilometre-Fiyat Grafiği')
plt.xlabel('Kilometre')
plt.ylabel('Fiyat')
plt.grid(True)
plt.legend()
plt.show()
```
